```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def ammonite_function(theta):
    a = 3.8
    b = 0.15
    c = 0.05
    d = 2
    return a * np.exp(b * theta) * (1 + c * np.sin(d * theta))

# Scaling Factor (a): This parameter determines the overall size of the spiral. 
# Increasing a will make the entire spiral larger, while decreasing it will make 
# the spiral smaller.

# Exponential Growth Rate (b): This parameter controls how quickly the spiral 
# expands as theta increases. A larger value of b will cause the spiral to grow 
# more rapidly, making it spread out faster. A smaller value will result in a more 
# tightly wound spiral.

# Amplitude of Sinusoidal Modulation (c): This parameter affects the amplitude 
# of the oscillations in the radius. A larger value of c will make the oscillations 
# more pronounced, creating a more wavy pattern. A smaller value will result in 
# less noticeable oscillations.

# Frequency of Sinusoidal Modulation (d): This parameter controls the frequency 
# of the oscillations. A larger value of d will increase the number of oscillations 
# within a given range of theta, making the spiral appear more wavy. A smaller value 
# will decrease the number of oscillations, resulting in a smoother spiral.

# Generate theta values
theta = np.linspace(0, 10 * np.pi, 1000)

# Calculate corresponding r values
r = ammonite_function(theta)

# Create a polar plot
plt.polar(theta, r)
plt.title("Ammonite Spiral")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Define the function
def ammonite_function(theta):
    a = 1.8
    b = 0.05
    c = 0.2
    d = 2
    return a * np.exp(b * theta) * (1 + c * np.sin(d * theta))

# Generate theta values
theta = np.linspace(0, 10 * np.pi, 1000)

# Calculate corresponding r values
r = ammonite_function(theta)

# Convert polar coordinates to Cartesian coordinates
x = r * np.cos(theta)
y = r * np.sin(theta)

# Create a figure and axis
fig, ax = plt.subplots()
ax.set_xlim(-max(r), max(r))
ax.set_ylim(-max(r), max(r))
line, = ax.plot([], [], lw=2)

# Initialization function
def init():
    line.set_data([], [])
    return line,

# Animation function
def animate(i):
    line.set_data(x[:i], y[:i])
    return line,

# Create the animation
ani = animation.FuncAnimation(fig, animate, init_func=init, frames=len(theta), interval=20, blit=True)

# Save the animation as a video file (optional)
ani.save('ammonite_spiral.mp4', writer='ffmpeg')

#plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def ammonite_function(theta):
    a = 3.8
    b = 0.15
    c = 0.05
    d = 2
    return a * np.exp(b * theta) * (1 + c * np.sin(d * theta))

# Generate theta values
theta = np.linspace(0, 10 * np.pi, 1000)

# Calculate corresponding r values
r = ammonite_function(theta)

# Create a polar plot with enhanced aesthetics
plt.figure(figsize=(10, 10))
ax = plt.subplot(111, projection='polar')
ax.plot(theta, r, color='teal', linewidth=2)

# Customize the plot aesthetics
ax.set_facecolor('#f7f7f7')  # Set background color
ax.grid(color='gray', linestyle='--', linewidth=0.5)  # Customize grid lines
ax.set_title("Ammonite Spiral", va='bottom', fontsize=20, color='teal')  # Set title with custom font size and color

# Remove the frame for a cleaner look
ax.spines['polar'].set_visible(False)

# Remove specific radial labels (50 to 400)
ax.set_yticklabels([])

plt.show()
```
